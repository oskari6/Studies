```python
import kagglehub
import pandas as pd
from sqlalchemy import create_engine

# Download latest version
# path = kagglehub.dataset_download("ahmedshahriarsakib/usa-real-estate-dataset")

# print("Path to dataset files:", path)

data = pd.read_csv("C:\\Users\\Oskari\\.cache\\kagglehub\\datasets\\ahmedshahriarsakib\\usa-real-estate-dataset\\versions\\25\\realtor-data.zip.csv")

data["brokered_by"] = data["brokered_by"].astype(str)
data["bed"] = data["bed"].fillna(0).astype(int)
data["bath"] = data["bath"].fillna(0).astype(int)
data['acre_lot'] = data['acre_lot'].fillna(0).astype(float)
data['street'] = data['street'].fillna(0).astype(float)
data["zip_code"] = data["zip_code"].fillna("").astype(str)
data["house_size"] = data["house_size"].fillna(0).astype(int)
data['prev_sold_date'] = pd.to_datetime(data['prev_sold_date'], errors='coerce').dt.date

engine = create_engine("mysql+mysqlconnector://oskari:password@localhost/cms_db")

try:
    data.to_sql("houses", con=engine, if_exists="replace", index=False, chunksize=1000)
except Exception as e:
    print("Error occurred: ", e)
```
